## Задание 3

```python
list_user = pd.read_csv('files/ratings1.csv')
list_user.head()
```

```python
list_user.groupby('userId').count().reset_index().sort_values('movieId', ascending=False).head()
# list_user.sort_values('count_mark', ascending = False)
# list_user
```

```python
count_user = list_user.value_counts('userId').reset_index().rename(columns={0:'count_mark'})
count_user = count_user[count_user['count_mark'] > 100]
count_user
```

```python
def len_life(user):
# for user in count_user['userId']:
#     print(user)
    maxtime = list_user[list_user['userId']==user]['timestamp'].max()
    mintime = list_user[list_user['userId']==user]['timestamp'].min()
    amplitude = maxtime - mintime
    return round(amplitude / 86400, 1)
```

Колонка lifetime - количество дней

```python
count_user['lifetime'] = count_user.userId.apply(len_life)
count_user
```
