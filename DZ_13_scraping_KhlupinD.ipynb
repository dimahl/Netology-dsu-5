```python
import json
report_ = json.loads(req.text);
report_
```

```python
report = pd.DataFrame()
for key, val in report_['breaches'].items():
#     print(key, val)
#     print('')
#     for key_1, val_1 in val:
#         print
    site = val['site']
    date = val['description'].split(',', 1)[0]
    description = val['description'].split(',', 1)[1]
    row = {'date': date, 'source': site, 'text':description}
    report = pd.concat([report, pd.DataFrame([row])])  
report.reset_index()
```
