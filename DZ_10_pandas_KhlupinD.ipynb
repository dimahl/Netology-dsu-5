##### Как можно сократить такое особенно категории 4,12,21 можно короче записать?

```python
filt_countr_cat = table_power[(
                                (table_power['country']=='Latvia') | (table_power['country']=='Lithuania') | 
                                (table_power['country']=='Estonia')
                                ) & (
                                (table_power['category']==4) | (table_power['category']==12) | (table_power['category']==21)
                             )]
filt_countr_cat
```

```python
filt_countr_cat_years_qua = filt_countr_cat[ (filt_countr_cat['year'].between(2005,2010)) & (filt_countr_cat['quantity']>0) ]
filt_countr_cat_years_qua
```

```python
Quantity_pribalt = filt_countr_cat_years_qua['quantity'].sum()
print(f'Суммарное потребление стран Прибалтики с заданным условием: {Quantity_pribalt}')
```
