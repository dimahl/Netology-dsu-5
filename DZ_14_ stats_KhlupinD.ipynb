## Задание 1. Загрузка данных

```python
horse = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv', names=range(1,29), na_values='?')
horse
```

## Задание 2. Первичное изучение данных

```python
horse.describe()
```

```python
print('Мода по наличию операций(1-да): ', horse['surgery?'].mode()[0])
print('Мода по судьбе лошади(1-умерла, 2-выжила, 3-усыпили): ', horse['outcome'].mode()[0])
print('Мода по наличию болей лошади(переменная легкая боль): ', horse['pain'].mode()[0])
```

```python
q1 = horse['rectal temperature'].quantile(0.25)
q3 = horse['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature')
remove_outliers
```

temperature of extremities
          - a subjective indication of peripheral circulation
          - possible values:
               1 = Normal
               2 = Warm
               3 = Cool
               4 = Cold
          - cool to cold extremities indicate possible shock
          - hot extremities should correlate with an elevated rectal temp.

```python
q1 = horse['pulse'].quantile(0.25)
q3 = horse['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

pulse
          - linear
          - the heart rate in beats per minute
          - is a reflection of the heart condition: 30 -40 is normal for adults
          - rare to have a lower than normal rate although athletic horses
            may have a rate of 20-25
          - animals with painful lesions or suffering from circulatory shock
            may have an elevated heart rate

#### Есть подозрение что есть ошибочные данные (ввода)  пульса, вообще все что выше сотни насторожило особенно 120, где 20 может быть нормой т.е. кажеться что единичку ввели ошибочно, особенно при низкой частоте дыханиии и отсутствии болии, т.е. у тех где pain=5 там еще может быть

```python
horse[horse['pulse'].between(120, 210, inclusive=True)].sort_values('pulse')
```

## Задание 3. Работа с пропусками

```python
(horse.isna().mean()*100).round(2)
```

удалю строки где оч много пропусков, а именно боллее 7

```python
horse1 = horse.dropna(thresh=7)
horse1
```

```python
(horse1.isna().mean()*100).round(2)
```

наличие болей заполню модой

```python
horse1['pain'].value_counts()
```

```python
horse1['pain'].fillna(horse1['pain'].mode()[0], inplace=True)
```

```python
(horse1.isna().mean()*100).round(2)
```

```python
horse2 = copy.copy(horse1)
horse2['pulse'].fillna(horse2.groupby(['outcome','pain'])['pulse'].transform('median'), inplace=True)
```

```python
(horse2.isna().mean()*100).round(2) 
```

#### Столбец частоты дыхания из-за разброса признан малополезным для исследования, удаляем

```python
horse2.drop(['respiratory rate'], axis=1, inplace=True)
```

```python
horse2.describe()
```

```python
q1 = horse2['rectal temperature'].quantile(0.05)
q3 = horse2['rectal temperature'].quantile(0.75)
print(q1, q3)
```

#### Чтобы заполнить t конечностей, разобъем выборку на дополнительные категории по ректальной температуре. Принимая во внимание, что лошади в больнице преобладают с повышенной температурой и нормальная температура 37,8 оказалась на первом квантиле, для группировки на 3 категории эмпереческим путем для нижней граници выбран 5ый процентиль, а для верхней граници - 75ый процентиль. 

```python
horse2.loc[horse2['rectal temperature'] < q1, 'rec_temp_cat'] = 1
horse2.loc[horse2['rectal temperature'] >= q1, 'rec_temp_cat'] = 2
horse2.loc[horse2['rectal temperature'] >= q3, 'rec_temp_cat'] = 3
```

```python
horse2
```

```python
horse2['temperature of extremities'].fillna(horse2.groupby('rec_temp_cat')['temperature of extremities'].transform('median'), inplace=True)
```

#### Проверяю нормально заполнилась категориальный столбец t.конечностей после медианы

```python
horse2.groupby(['temperature of extremities']).count()
```

```python
(horse2.isna().mean()*100).round(2) 
```

```python
horse2['rectal temperature'].fillna(horse2.groupby('temperature of extremities')['rectal temperature'].transform('median'), inplace=True)
```

```python
horse2.drop(['rec_temp_cat'], axis=1, inplace=True)
```

```python
(horse2.isna().mean()*100).round(2) 
```
