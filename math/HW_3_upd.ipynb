### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


## Уменьшить объект в два раза

````python
small_A = A / 2

x = small_A[:,0]
y = small_A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

## Повернуть на 130 градусов

````python
fi = 130
rad_fi = np.radians(fi)

orto_matrix = np.array([[np.cos(rad_fi), -np.sin(rad_fi), 0],
                        [np.sin(rad_fi),  np.cos(rad_fi), 0],
                        [0,0,1]])

rot_matrix = A@orto_matrix

x = rot_matrix[:,0]
y = rot_matrix[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

## Отразить объект относительно прямой y=x

````python
t = [[0, 1, 0], [1, 0, 0], [0, 0, 1]]


mirr_matrix = A@t

x = mirr_matrix[:,0]
y = mirr_matrix[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

вспомогательная матрица не очень помогла, но сделал в два действия, повернул на 135 и отзеркалил относительно начало координат

````python
fi = 135
rad_fi = np.radians(fi)

orto_matrix = np.array([[np.cos(rad_fi), -np.sin(rad_fi), 0],
                        [np.sin(rad_fi),  np.cos(rad_fi), 0],
                        [0,0,1]])

rot_matrix = A@orto_matrix

x = rot_matrix[:,0]
y = rot_matrix[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
t = [[-1, 0, 0], [0, -1, 0], [0, 0, -1]]

rot_matrix_yx = rot_matrix@t

x = rot_matrix_yx[:,0]
y = rot_matrix_yx[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````
